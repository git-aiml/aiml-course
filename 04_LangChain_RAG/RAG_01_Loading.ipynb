# PDF Loader

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader_pdf = PyPDFLoader("Introduction_to_Data_and_Data_Science.pdf")
```

```python
pages_pdf = loader_pdf.load();
```

```python
len(pages_pdf)
#pages_pdf
```

```python
import copy
```

```python
pages_pdf_cut = copy.deepcopy(pages_pdf)
```

```python
for i in pages_pdf_cut:
    i.page_content = ' '.join(i.page_content.split())
```

```python
len(pages_pdf_cut)
#pages_pdf_cut
```

```python
import tiktoken
encoding = tiktoken.get_encoding("o200k_base")  # supports GPT-4, GPT-4o, GPT-5-mini etc.
```

```python
original_tokens = sum(len(encoding.encode(i.page_content)) for i in pages_pdf)
print("Total tokens :", original_tokens)
```

```python
reduced_tokens = sum(len(encoding.encode(i.page_content)) for i in pages_pdf_cut)
print("Total tokens after Removing newline:", reduced_tokens)
```

```python
percent_reduction = ((original_tokens - reduced_tokens) / original_tokens) * 100
print(percent_reduction)
```

# Document Loader

```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
loader_docx = Docx2txtLoader("Introduction_to_Data_and_Data_Science.docx")
```

```python
pages_docx = loader_docx.load()
```

```python
len(pages_docx)
#pages_docx
```

```python
for i in range(len(pages_docx)):
    pages_docx[i].page_content = ' '.join(pages_docx[i].page_content.split())
```

```python
len(pages_docx[0].page_content)
```

```python
char_splitter = CharacterTextSplitter(separator = "",
                                      chunk_size = 500,
                                      chunk_overlap = 50)
```

```python
pages_char_split = char_splitter.split_documents(pages_docx)
pages_char_split[0].page_content
```
