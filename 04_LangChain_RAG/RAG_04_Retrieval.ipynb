```python
import sys
sys.path.append("..")
import sks_config
# sks_config.SKS_OPENAI_API_KEY    
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
embedding = OpenAIEmbeddings(api_key = sks_config.SKS_OPENAI_API_KEY, 
                             model = "text-embedding-ada-002")
```

```python
vectorstore = Chroma(persist_directory = "./sks-chromadb",
                     embedding_function = embedding)
```

```python
added_document = Document(page_content = "In terms of predictive analytics, EViews is mostly used for working with econometric time-series models, and Stata—for academic statistical and econometric research, where techniques like regression, cluster, and factor analysis are constantly applied. As a final note, remember the following. Should you have the relevant business and theoretical knowledge, learning a software tool is relatively easy as opposed to learning a programming language",
                         metadata = {"Course Title": "Introduction to Data and Data Science",
                                      "Lecture Title": "As you have probably guessed, analytics generally"})
```

```python
vectorstore.add_documents([added_document])
```

```python
question = "Which programig language the data cientist use?"
```

```python
retrived_docs = vectorstore.similarity_search(query = question,
                                              k = 5)
```

```python
retrived_docs
```

```python
for i in retrived_docs:
    print(f"page Content: {i.page_content}\n -----------------\nLecture Title: {i.metadata['Lecture Title']}\n")
```

```python
question = "Which softwares the data cientist use?"
```

```python
retrived_docs = vectorstore.similarity_search(
    query = question,
    k = 3)
```

```python
for i in retrived_docs:
    print(f"page Content: {i.page_content}\n -----------------\nLecture Title: {i.metadata['Lecture Title']}\n")
```

```python
retrived_docs = vectorstore.max_marginal_relevance_search(
    query = question,
    k = 3,
    lambda_mult = 0.1,
    filter = {"Lecture Title": "As you have probably guessed, analytics generally"})
```

```python
for i in retrived_docs:
    print(f"page Content: {i.page_content}\n -----------------\nLecture Title: {i.metadata['Lecture Title']}\n")
```

```python
retriever = vectorstore.as_retriever(search_type = "mmr", 
                                     search_kwargs = {'k': 3, 
                                                      'lambda_mult': 0.7})
```

```python
retriever
```

```python
question = "Which softwares the data cientist use?"
```

```python
retrieved_docs = retriever.invoke(question)
```
