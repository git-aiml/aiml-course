# Batching & Streaming

```python
import sys
sys.path.append("..")
import sks_config
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
```

```python
chat_template = ChatPromptTemplate([
    ('human',
     "I have recently adopted a {pet} which is a {breed}. Please suggest training tips.")])
```

```python
chat = ChatOpenAI(api_key = sks_config.SKS_OPENAI_API_KEY,
                  model_name = 'gpt-4o-mini',
                  model_kwargs = {'seed': 365},
                  temperature = 0,
                  max_tokens = 100)         
```

```python
chain = chat_template | chat
```

```python
chain.invoke({'pet': 'dog', 'breed': 'bulldog'})
```

```python
%%time
chain.batch([{'pet': 'dog', 'breed': 'bulldog'}, {'pet': 'dragon', 'breed': 'night fury'}])
```

```python
%%time
chain.invoke({'pet': 'dog', 'breed': 'bulldog'})
```

```python
%%time
chain.invoke({'pet': 'dragon', 'breed': 'night fury'})
```

```python
response = chain.stream({'pet': 'dog', 'breed': 'bulldog'})
```

```python
next(response)
```

```python
for i in response:
    print(i.content, end = '')
```
