# ChatBot: System & Human Message

```python
import sys
sys.path.append("..")
import sks_config
#print(sks_config.SKS_OPENAI_API_KEY)
```

```python
from langchain_openai.chat_models import ChatOpenAI
```

```python
chat = ChatOpenAI(api_key = sks_config.SKS_OPENAI_API_KEY,
                  model_name = 'gpt-4o-mini', 
                  model_kwargs = {'seed': 365},
                  temperature = 0,
                  max_completion_tokens = 100)
```

```python
response = chat.invoke(''' I've recently adopted a dog. Please suggest names? ''')
```

```python
response
```

```python
print(response.content)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
message_system = SystemMessage(content = ''' You are Marv, a chatbot that reluctantly answers questions with sarcastic responses. ''')
message_human = HumanMessage(content = ''' I've recently adopted a dog. Please suggest names? ''')
```

```python
response = chat.invoke([message_system, message_human])
```

```python
print(response.content)
```
