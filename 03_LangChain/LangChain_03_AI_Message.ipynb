# ChatBot: AI Message

```python
import sys
sys.path.append("..")
import sks_config
#print(sks_config.SKS_OPENAI_API_KEY)
```

```python
import openai
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage
```

```python
chat = ChatOpenAI(api_key = sks_config.SKS_OPENAI_API_KEY,
                  model_name = 'gpt-4o-mini',
                  model_kwargs = {'seed': 365},
                  temperature = 0,
                  max_tokens = 100
                 )
```

```python
message_human = HumanMessage(content = ''' I've recently adopted a dog. Please suggest names? ''')
message_ai = AIMessage(content = ''' Oh, absolutely. Because nothing screams "I'm responsible pet owner" 
like asking a chatbot to name your new furball. How about "Bark Twain" (if it's a literary houns)? ''')

message_human_cat = HumanMessage(content = ''' I've recently adopted a cat. Please suggest names? ''')
message_ai_cat = AIMessage(content = ''' Oh, absolutely. Because nothing screams "I'm unique and creative individual" 
like asking a chatbot to name your cat. How about "Furry McFurFace", "Sir Meowsalot", or "Catastrophe"? ''')

message_human_fish = HumanMessage(content = ''' I've recently adopted a fish. Please suggest names? ''')
```

```python
response = chat.invoke([message_human, message_ai, message_human_cat, message_ai_cat, message_human_fish])
```
