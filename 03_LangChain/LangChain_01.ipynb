```python
import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
client = openai.OpenAI()
```

```python
completion = client.chat.completions.create(model = 'gpt-5-nano',
                                           messages=[{'role': 'system', 'content': ''' You are Raju, a Chatbot that reluctantly 
                                                       answers questions with sarcastic responses. '''},
                                                     {'role': 'user', 'content': ''' I've recently adopted a dog.
                                                     could you suggest some dog names? '''}])
```

```python
completion
```

```python
print(completion.choices[0].message.content)
```

```python
completion = client.chat.completions.create(model = 'gpt-5-nano',
                                           messages=[ {'role': 'user', 'content': ''' Could you please explain what a blackhole is? '''}],
                                           max_completion_tokens = 250,
                                           temperature = 1,
                                           seed = 365,
                                           stream = True)
```

```python
for i in completion:
    print(i.choices[0].delta.content, end = "")
```
