```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence = "it was too far to go to the shop and he did not want her to walk"
sentence_no_stopwords = ' '.join([word for word in sentence.split() if word not in en_stopwords])
print(sentence_no_stopwords)
```

```python
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
sentence = "This is the to tell that, I have made my decision to start a tech company based on AI/ML. I am not sure about the domain and problem but I will figurel it out within a month, by Nov-2025, and start working on it. Thats a determination. There is no substitute to hardwork and determination"
words = word_tokenize(sentence)
print(words)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# unigrams: n=1
unigrams = (pd.Series(nltk.ngrams(words, 1)).value_counts())
print(unigrams[:10])
```

```python
unigrams[:10].sort_values().plot.barh(color="lightsalmon", width=.9, figsize=(12, 8))
plt.title("10 Most Frequently Occuring Unigrams")
```

```python
bigrams = (pd.Series(nltk.ngrams(words, 2)).value_counts())
print(bigrams[:10])
```

```python
bigrams[:10].sort_values().plot.barh(color="darksalmon", width=.9, figsize=(12, 8))
plt.title("10 Most Frequently Occuring Unigrams")
```
