```python
import sys
sys.path.append("..")
import sks_config
#print(sks_config.SKS_OPENAI_API_KEY)
```

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

import langchain
print("langchain version: ", langchain.__version__)

#
#The simplest and most direct fix for this issue is to install the langchain-classic package, which contains the memory 
#and chain components from the pre-1.0 architecture.
#
#from langchain_community.memory import ConversationBufferMemory
#from langchain.memory import ConversationBufferMemory
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.llms import OpenAI
from langchain_classic.chains import ConversationalRetrievalChain
from langchain_classic.chat_models import ChatOpenAI

import sys
print(sys.executable)
```

```python
url = "https://365datascience.com/upcomig-courses"
```

```python
loader = WebBaseLoader(url)
```

```python
raw_documents = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(raw_documents)
```

```python
embeddings = OpenAIEmbeddings(openai_api_key = sks_config.SKS_OPENAI_API_KEY)
```

```python
vectorstore = FAISS.from_documents(documents, embeddings)
```

```python
memory = ConversationBufferMemory(memory_key = "chat_history", return_messages = True)
```

```python
qa = ConversationalRetrievalChain.from_llm(ChatOpenAI(openai_api_key=sks_config.SKS_OPENAI_API_KEY,
                                                      model="gpt-5-nano",
                                                      temperature=0),
                                           vectorstore.as_retriever(),
                                           memory=memory)
```

```python
query ="What is the next course uploaded on 365DataScience Platform?"
result = qa({"question": query})
```
