```python
pc = Pinecone(api_key = sks_config.SKS_PINECONE_API_KEY)
```

```python
pc.list_indexes()
```

```python
index_name = "sks-index-3"
dimension = 3
metric = "cosine"
```

```python
if index_name in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name)
    print(f"{index_name} succesfully deleted.")
else:
     print(f"{index_name} not in index list.")
```

```python
pc.list_indexes()
```

```python
pc.create_index(
    name = index_name, 
    dimension = dimension, 
    metric = metric, 
    spec = ServerlessSpec(
        cloud = "aws", 
        region = "us-east-1")
    )
```
