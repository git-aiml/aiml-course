## Creating new Index

```python
index_name = "sks-index-2"
dimension = 3
metric = "cosine"
```

```python
if index_name in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name)
    print(f"{index_name} successfully deleted.")
else:
    print(f"{index_name} not in index list.")
```

```python
pc.create_index(
    name = index_name,
    dimension = dimension,
    metric = metric,
    spec = ServerlessSpec(
        cloud = "aws",
        region = "us-east-1")
)
```

```python
pc.list_indexes()
```

```python
index = pc.Index(name = index_name)
```

```python
index.upsert([
    ("Dog", [4., 0., 1.]),
    ("Cat", [4., 0., 1.]),
    ("Chicken", [2., 2., 1.]),
    ("Mantis", [6., 2., 3.]),
    ("Elephant", [4., 0., 1.])
])
```
